```python
import pandas as pd
import matplotlib.pylab as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import *
import numpy as np
```

```python
data = pd.read_csv('datasets/Bike-Sharing-Dataset/day.csv')
```

```python
%matplotlib inline
threedee = plt.figure().gca(projection='3d')
threedee.scatter(data.hum, data.temp, data.cnt)
threedee.set_xlabel('hum')
threedee.set_ylabel('temp')
threedee.set_zlabel('Count')
plt.show()
```

```python
l = LinearRegression(fit_intercept=True, normalize=False, copy_X=True, n_jobs=1)
```

```python
l.fit(data[['temp', 'hum']].values.reshape(-1,2), data.cnt)
```

```python
temp_values = np.arange(0, .41, .05)
hum_values = np.arange(0, 1, .1)
xy_grid = np.meshgrid(temp_values, hum_values)
temp_hum_combinations = np.array(xy_grid).T.reshape(-1,2)
predictions = l.predict(temp_hum_combinations)
#predictions.reshape(-1,2)
#print(xy_grid[0])
#print(xy_grid[1])
X = np.delete(temp_hum_combinations, 1, 1)
Y = np.delete(temp_hum_combinations, 0, 1)
xx, yy = np.meshgrid(X, Y)
```

```python
predictions_plane = plt.figure().gca(projection='3d')
predictions_plane.plot_surface(yy, xx, predictions) #switch x/y because it's easier to see
predictions_plane.set_xlabel('hum')
predictions_plane.set_ylabel('temp')
predictions_plane.set_zlabel('Count')
threedee = plt.figure().gca(projection='3d')
threedee.scatter(data.hum, data.temp, data.cnt)
threedee.set_xlabel('hum')
threedee.set_ylabel('temp')
threedee.set_zlabel('Count')
plt.show()
```
