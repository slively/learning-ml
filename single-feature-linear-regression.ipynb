```python
import pandas as pd
from matplotlib.pylab import *
from sklearn.linear_model import *
import numpy as np
```

```python
data = pd.read_csv('datasets/Bike-Sharing-Dataset/day.csv')
```

```python
%matplotlib inline
scatter(data.temp, data.cnt)
```

```python
l = LinearRegression(fit_intercept=True, normalize=False, copy_X=True, n_jobs=1)
```

```python
l.fit(data.temp.values.reshape(-1, 1), data.cnt)
```

```python
x = np.arange(0, 1, .1)
```

```python
y = l.predict(x.reshape(-1, 1))
```

```python
plot(x,y)
scatter(data.temp, data.cnt)
```
