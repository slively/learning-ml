```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.datasets import make_moons, make_circles, make_classification
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.model_selection import cross_val_score
```

```python
data = pd.read_csv('datasets/car.data')
data[0:10]
```

```python
buying_label_encoder = LabelEncoder()
buying_label_encoder.fit(['vhigh', 'high', 'med', 'low'])
processed_buying = buying_label_encoder.transform(data.buying)
```

```python
maint_encoder = buying_label_encoder
processed_maint = maint_encoder.transform(data.maint)
```

```python
doors_label_encoder = LabelEncoder()
doors_label_encoder.fit(['2', '3', '4', '5more'])
processed_doors = doors_label_encoder.transform(data.doors)
```

```python
persons_label_encoder = LabelEncoder()
persons_label_encoder.fit(['2', '4', 'more'])
processed_persons_labels = persons_label_encoder.transform(data.persons)
```

```python
lug_boot_label_encoder = LabelEncoder()
lug_boot_label_encoder.fit(['small', 'med', 'big'])
processed_lug_boots = lug_boot_label_encoder.transform(data.lug_boot)
```

```python
safety_label_encoder = LabelEncoder()
safety_label_encoder.fit(['low', 'med', 'high'])
processed_safety = safety_label_encoder.transform(data.safety)
```

```python
classification_label_encoder = LabelEncoder()
classification_label_encoder.fit(['unacc', 'acc', 'good', 'vgood'])
processed_classifications = classification_label_encoder.transform(data.classification)
```

```python
processed_data = pd.DataFrame({
    'buying': processed_buying,
    'maint': processed_maint, 
    'doors': processed_doors,
    'persons': processed_persons_labels, 
    'lug_boots': processed_lug_boots,
    'safety': processed_safety
})
processed_data[0:10]
```

```python
classifiers = {
    "Nearest Neighbors": KNeighborsClassifier(),
    #"Linear SVM": SVC(kernel="linear", C=0.025),
    #"RBF SVM": SVC(gamma=2, C=1),
    #"Gaussian Process": GaussianProcessClassifier(1.0 * RBF(1.0), warm_start=True),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    #"Neural Net": MLPClassifier(alpha=1),
    #"AdaBoost": AdaBoostClassifier(),
    "Naive Bayes": GaussianNB(),
    #"QDA": QuadraticDiscriminantAnalysis()
}
```

```python
for name, clf in classifiers.items():
    scores = cross_val_score(clf, processed_data, processed_classifications, cv=10)
    '''
    for i in np.arange(10):
        train_data, test_data, train_classifications, test_classifications = train_test_split(processed_data, processed_classifications)
        clf.fit(train_data, train_classifications)
        score = clf.score(train_data, train_classifications)
        print('train - {0}: ~{1}%'.format(name, int(score*100)))
        score = clf.score(test_data, test_classifications)
        print('test - {0}: ~{1}%'.format(name, int(score*100)))
        scores.append(score)
        '''
    print('{0} average: ~{1}% (+/-{2})'.format(name, scores.mean(), scores.std() * 2))
```
